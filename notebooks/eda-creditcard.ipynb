```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

```python
credit_df = pd.read_csv("../data/raw/creditcard.csv")
```

```python
credit_df.head()

credit_df.shape

credit_df.info()

credit_df.describe()
```

```python
credit_df.isnull().sum()
```

```python
credit_df.duplicated().sum()
```

### Observation

A total of 1081 duplicate transactions were identified and removed to prevent bias in model training and evaluation.

```python
credit_df = credit_df.drop_duplicates()
```

```python
credit_df["Class"].value_counts()

credit_df["Class"].value_counts(normalize=True) * 100
```

### Observation

The dataset is highly imbalanced, with fraudulent transactions accounting for only 0.17% of all transactions. This makes accuracy an unreliable evaluation metric and justifies the use of resampling techniques such as SMOTE during model training.

```python
sns.countplot(x="Class", data=credit_df)

plt.title("Class Distribution")
plt.show()
```

```python
sns.histplot(credit_df["Amount"], bins=50)

plt.title("Distribution of Transaction Amount")
plt.show()
```

### Observation

Transaction amounts are highly right-skewed. Most transactions involve relatively small amounts, while a small number of transactions involve very large values.

```python
sns.histplot(credit_df["Time"], bins=50)

plt.title("Distribution of Transaction Time")
plt.show()
```

```python
sns.boxplot(
    x="Class",
    y="Amount",
    data=credit_df
)

plt.title("Transaction Amount by Class")
plt.show()
```

```python
plt.figure(figsize=(20, 15))

sns.heatmap(
    credit_df.corr(),
    cmap="coolwarm",
    center=0
)

plt.title("Correlation Matrix")
plt.show()
```

```python
credit_df.corr()["Class"].sort_values()
```

### Observation

Among the PCA-transformed variables, V17, V14, V12, V10, and V16 exhibit the strongest relationships with fraudulent transactions and may play an important role in classification performance.
